```python
import pandas as pd
import sqlite3 as sq3
```

```python
#####Question 1

TableData = pd.read_csv("Wk4TableData.csv", dtype = str)

for i in range(0,TableData.shape[0]):
    for j in range(0,TableData.shape[1]):
        if pd.isnull(TableData.iloc[[i], [j]]).iloc[0,0]:
            TableData.iloc[i, j] = TableData.iloc[(i-1), j]
            
TableData.drop_duplicates()
TableData.dropna()
```

```python
StudentData = TableData.iloc[: , 0:4]
StudentData = StudentData.drop_duplicates()
StudentData.set_index('Student Number', inplace=True)
StudentData
```

```python
StudentCourse = TableData.iloc[: , [0,4]]
StudentCourse = StudentCourse.drop_duplicates()
StudentCourse.set_index(['Student Number', 'Course Number'], inplace=True)
StudentCourse
```

```python
CourseData = TableData.iloc[: , 4:7]
CourseData = CourseData.drop_duplicates()
CourseData.set_index('Course Number', inplace=True)
CourseData
```

```python
TeacherData = TableData.iloc[: , 6:]
TeacherData = TeacherData.drop_duplicates()
TeacherData.set_index('TeacherID', inplace=True)
TeacherData
```

```python
########Question 2

sq3_file='StudentCourseInfo.db'
con=sq3.connect(sq3_file)

StudentData.to_sql('StudentData', con, if_exists='replace')
StudentCourse.to_sql('StudentCourse', con, if_exists='replace')
CourseData.to_sql('CourseData', con, if_exists='replace')
TeacherData.to_sql('TeacherData', con, if_exists='replace')

#Note - The primary key is not set because there does not seem to be a method to easily do this:
#https://stackoverflow.com/questions/39407254/how-to-set-the-primary-key-when-writing-a-pandas-dataframe-to-a-sqlite-database?rq=1

con.commit()
con.close()
```

```python
#Creating a database with a primary key

sq3_file='RandomDB.db'
con=sq3.connect(sq3_file)

with con:
    con.execute('CREATE TABLE Cats (catID INT PRIMARY KEY, name TEXT, ownerID INT)')
    data=[
        (4812, 'Momo', '2285'),
        (4813, 'Fluffy', '2287'),
        (4814, 'Zippy', '2285'),
        (4815, 'Toes', '2289'),
        (4816, 'Rusty', '2288'),
        (4817, 'Gatsby', '2288')
        ]
    for d in data:
        con.execute('INSERT INTO Cats VALUES (?,? ,?)', d)

with con:
    con.execute('CREATE TABLE Owners (ownerID INT PRIMARY KEY, name TEXT)')
    data=[
        (2285, 'Ida'),
        (2286, 'Kelsey'),
        (2287, 'Anya'),
        (2288, 'Brittany'),
        (2289, 'Joe')
        ]
    for d in data:
        con.execute('INSERT INTO Owners VALUES (?,?)', d)

con.commit()
con.close()
```
